```
import numpy as np
import seaborn as sns
import numpy.random as rd
import matplotlib.pyplot as plt
import math
from IPython.display import display, Math
import scipy.stats as st
```

Пусть $\xi_i \in \mathcal{Bin}(n, p)$. Покажите, что при $np \to \lambda $ $n \to \inf$  биномиальное распределение $Bin(n, \frac{\lambda}{n})$ стремится к распределению Пуассона $Pois(\lambda)$. Чтобы продемонстрировать сходимость построим несколько графиков при разных значениях n.




```
# np -> lam , --->> p = lam / n
cmap = plt.get_cmap("tab10").colors

plt.figure(figsize=(15, 15))

lam = 21

for k, r in enumerate([10, 50, 100, 500, 1000, 5000, 10000, 50000, 100000]):
    p = lam / r
    c = [st.binom.pmf(i, r, p) for i in range(30)] # binom.pmf(x, n, p)

    plt.subplot(3, 3, k+1)
    plt.tight_layout()
    y = [st.poisson.pmf(i, lam) for i in range(30)]
    plt.plot(list(range(30)), y, '.-', c='k', label="Poiss")
    plt.scatter(list(range(30)), c, marker='s', color=cmap[k], label="n = " + str(r))
    #plt.xticks(x, x)
    plt.xlabel("k")
    plt.ylabel(r"$P(\xi _n=k)$")
    plt.legend()
plt.show()
```

Вывод: по графикам видно, что при достаточно больших n, начиная с n=500 распределения сходятся.
